```kotlin
import com.tamj0rd2.ktcheck.Gens
import com.tamj0rd2.ktcheck.full

// visualises the distribution of samples from a generator. useful when trying to understand the consequences of changes to edge case probability, or overall generation.
val gen = Gens.int(1..100)
val samples = gen.samples(1).take(1_000).toList()
val grouped = samples.groupingBy { it }.eachCount()
plot {
    bars {
        x(grouped.keys, "number")
        y(grouped.values, "count")
    }
}
```

```kotlin
import com.tamj0rd2.ktcheck.Gens
import com.tamj0rd2.ktcheck.full

// visualises the distribution of combined samples from a generator. useful when trying to understand the consequences of changes to edge case probability, or overall generation.
val gen = Gens.zip(Gens.int(1..5), Gens.int(1..5)).map { (a, b) -> "${a}_${b}" }
val samples = gen.samples(1).take(10_000).toList().sorted()
val grouped = samples.groupingBy { it }.eachCount()
plot {
    bars {
        x(grouped.keys, "pair")
        y(grouped.values, "count")
    }
}
```
